```
datelist = []
day=22

for month in range (1,6):
  if month==1 or month==3 or month==5:
    for day in range (1,10):
      datelist.append('0'+str(month)+'-0'+str(day)+'-2020')
    for day in range (10,32):
      datelist.append('0'+str(month)+'-'+str(day)+'-2020')
  elif month==2:
    for day in range (1,10):
      datelist.append('0'+str(month)+'-0'+str(day)+'-2020')
    for day in range (10,30):
      datelist.append('0'+str(month)+'-'+str(day)+'-2020')
  elif month==4:
    for day in range (1,10):
      datelist.append('0'+str(month)+'-0'+str(day)+'-2020')
    for day in range (10,31):
      datelist.append('0'+str(month)+'-'+str(day)+'-2020')

print(datelist)
del datelist[:22]
print(datelist)
```

```
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/05-15-2020.csv' 
df = pd.read_csv(url).fillna(0)

for date in datelist:
  url_loop = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/'+date+'.csv'
  df_loop = pd.read_csv(url_loop).fillna(0)
  df.append(df_loop, ignore_index=True)
```
